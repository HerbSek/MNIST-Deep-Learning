```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Dataset
from torchvision import models
import pandas as pd
```

```python
class ImageDataset(Dataset):
    def __init__(self, image_tensor, labels_df):
        self.images = image_tensor
        self.labels = torch.tensor(labels_df, dtype = torch.long)
    def __len__(self):
        return len(self.images)
    def __getitem__(self, idx):
        image = self.images[idx].unsqueeze(0)
        label = self.labels[idx]
        return image,label
```

```python
### create dataset

image_tensor = x_data[:63000]
labels_df = y_data[:63000]

#Test_data
test_tensors = x_data[63000:]
```

```python
dataset = ImageDataset(image_tensor, labels_df)
# dataloader for batching below
dataloader = DataLoader(dataset, batch_size = 70, shuffle = True) 
```

```python
class cnn(nn.Module):
    def __init__(self):
        super(cnn, self).__init__()
        self.model = nn.Sequential(
               nn.Conv2d(1,16,3,1,1),
               nn.ReLU(),
               nn.MaxPool2d(2,2),

               nn.Conv2d(16, 32, 3,1,1),
               nn.ReLU(),
               nn.MaxPool2d(2,2),
              
           )
        self.linear1 =  nn.Linear(32*7*7, 128)
        self.linear2 = nn.ReLU()
        self.linear3 = nn.Linear(128,10)
        
    def forward(self, x):
        x = self.model(x)
        x = x.view(-1, 32*7*7)
        x = self.linear1(x)
        x = self.linear2(x)
        x = self.linear3(x)
        return x
```

```python
image_tensor[0].shape
```

```python
test_tensor =  image_tensor[0].unsqueeze(0).unsqueeze(0)
test_tensor.shape
```

```python
model = cnn()
```

```python
model
```

```python
# Training model
import torch.optim as optim
loss_function = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr = 0.001)
```

```python
epochs = 3
for epoch in range(epochs):
    running_loss = 0.0
    for images, labels in dataloader:
        optimizer.zero_grad()
        
        output = model(images)
        loss = loss_function(output, labels)
        loss.backward()
        optimizer.step()
    print(f" Epoch {epoch+1}, loss: {loss} ")

print("Finished training")        
```
